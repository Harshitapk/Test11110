```python
import pandas as pd
import requests
import os
import json
# Read CSV file into a DataFrame
#os.chdir('/Users/harshita/Downloads/Final_project/notebooks')
All_Cities_df = pd.read_csv("../data/All_Cities.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for 'Rating'
plt.figure(figsize=(8, 6))
sns.histplot(All_Cities_df['Rating'], kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.savefig("../images/Plot1.png")
plt.show()

# Box plot for 'Rating'
plt.figure(figsize=(8, 6))
sns.boxplot(x=All_Cities_df['Closed'], y=All_Cities_df['Rating'])
plt.title('Rating Distribution for Closed and Open Restaurants')
plt.xlabel('Closed')
plt.ylabel('Rating')
plt.savefig("../images/Plot2.png")
plt.show()


# Scatter plot for 'Rating' vs. 'Reviews'
plt.figure(figsize=(8, 6))
sns.scatterplot(x=All_Cities_df['Rating'], y=All_Cities_df['Reviews'])
plt.title('Rating vs. Reviews')
plt.xlabel('Rating')
plt.ylabel('Reviews')
plt.savefig("../images/Plot3.png")
plt.show()


```

```python
# Filter top 10 categories based on reviews
top_categories = All_Cities_df.dropna(subset=['Categories']).nlargest(30, 'Reviews')

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Reviews', y='Categories', data=top_categories, palette='viridis')
plt.xlabel('Reviews')
plt.ylabel('Categories')
plt.title('Top 30 Categories Based on Reviews')
plt.savefig("../images/Plot4.png")
plt.show()
```

```python
# Filter top 10 categories based on reviews
top_categories = All_Cities_df.dropna(subset=['Cuisine Types']).nlargest(30, 'Reviews')

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Reviews', y='Cuisine Types', data=top_categories, palette='viridis')
plt.xlabel('Reviews')
plt.ylabel('Cuisine Types')
plt.title('Top 30 Cuisine Types Based on Reviews')
plt.savefig("../images/Plot5.png")
plt.show()
```

```python
# Filter top 10 categories based on reviews
top_categories = All_Cities_df.dropna(subset=['Nationalities']).nlargest(30, 'Reviews')

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Reviews', y='Nationalities', data=top_categories)
plt.xlabel('Reviews')
plt.ylabel('Nationalities')
plt.title('Top 30 Nationalities Based on Reviews')
plt.savefig("../images/Plot6.png")
plt.show()
```

```python
# Group by 'Nationalities' and calculate the average rating for each category
average_ratings = All_Cities_df.groupby('Nationalities')['Rating'].mean().reset_index()

# Sort by average rating and select the top 30 categories
top_categories = average_ratings.nlargest(30, 'Rating')

# Create a bar plot using Seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='Rating', y='Nationalities', data=top_categories, palette='viridis')
plt.xlabel('Average Rating')
plt.ylabel('Categories')
plt.title('Top 30 Categories Based on Average Rating Across Cities')
plt.savefig("../images/Plot7.png")
plt.show()
```
