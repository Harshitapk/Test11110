```python
import requests
import json
import pandas as pd

YELP_API_KEY = YELP_API_KEY
API_ENDPOINT = 'https://api.yelp.com/v3/businesses/search'

# Yelp API endpoint URL for exploring restaurants
yelp_url = 'https://api.yelp.com/v3/businesses/search'
yelp_headers = {
    "Authorization": f"Bearer {YELP_API_KEY}"
}

venues_data_list = []

yelp_params = {
    'location': 'Toronto',
    'radius': 5000,  # Radius for restaurant search (adjust as needed)
    'categories': 'restaurants',
    'limit': 100,  # Number of restaurants to retrieve (adjust as needed)
}

yelp_response = requests.get(API_ENDPOINT, params=yelp_params, headers=yelp_headers)

# Check if the Yelp response is successful
if yelp_response.status_code == 200:
    yelp_data = json.loads(yelp_response.text)
    for business in yelp_data.get('businesses', []):
        categories = [category.get('title') for category in business.get('categories', [])]
        address1 = business.get('location', {}).get('address1')
        city = business.get('location', {}).get('city')
        if address1 and city:
            address = ', '.join([address1, city])
        else:
            address = 'N/A'  # You can set a default value or handle it in your own way
        restaurant_details = {
            'Name': business.get('name'),
            'Categories': ', '.join(categories),
            'Rating': business.get('rating'),
            'Address': address,
            'Res_Latitude': business.get('coordinates', {}).get('latitude'),
            'Res_Longitude': business.get('coordinates', {}).get('longitude'),
            'Reviews': business.get('review_count'),
            'Distance': business.get('distance'),
            'Phone': business.get('phone'),
            'Closed': business.get('is_closed'),
            'Price': business.get('price')
        }
        venues_data_list.append(restaurant_details)

# Create a DataFrame from the restaurant data list
restaurants_df = pd.DataFrame(venues_data_list)

# Print the DataFrame
print(restaurants_df)
```

```python
import pandas as pd
Toronto1_df = pd.read_csv('../data/Toronto.csv')
Toronto1_df.head()
```

```python
Toronto1_df['Price'].fillna(method='ffill', inplace=True)
columns_to_fill = ['Nationalities','Phone','Categories']
Toronto1_df[columns_to_fill] = Toronto1_df[columns_to_fill].fillna('NA')


import pandas as pd
import numpy as np
def fill_cuisine(row):
    if pd.notnull(row['Cuisine Types']):
        return row['Cuisine Types']
    categories = row['Categories'].split(', ')
    if len(categories) > 0:
        return categories[0]
    return np.nan

# Apply the function to fill 'Cuisine Types'
Toronto1_df['Cuisine Types'] = Toronto1_df.apply(fill_cuisine, axis=1)

Toronto1_df['Phone'] = pd.to_numeric(Toronto1_df['Phone'], errors='coerce').fillna(np.nan).astype('Int64').astype(str)

Toronto1_df['City']='Toronto'
# Print the updated DataFrame
Toronto1_df.head()
```

```python
Toronto1_df['Res_Latitude'] = Toronto1_df['Res_Latitude'].round(6)
Toronto1_df['Res_Longitude'] = Toronto1_df['Res_Longitude'].round(6)
Toronto1_df['Distance']=Toronto1_df['Distance'].round(2)
```
